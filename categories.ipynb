```python
import os
import dotenv
import zeep
import json
from tap_tradera.client import TraderaClient
```

```python
dotenv.load_dotenv(".env")
```

```python
client = TraderaClient(
    app_id=int(os.getenv("TAP_TRADERA_APP_ID")), app_key=os.getenv("TAP_TRADERA_APP_KEY")
)
```

```python
categories = client.get_categories()
categories
```

```python
from genson import SchemaBuilder

builder = SchemaBuilder()
for category in categories:
    builder.add_object(category)
json_schema = builder.to_json(indent=2)
print(json_schema)
```
